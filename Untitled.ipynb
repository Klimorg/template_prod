## Opening and Resizing

`interpolation=cv2.INTER_LINEAR` in cv2 is *almost* equivalent to `PIL.Image.BILINEAR`.

https://stackoverflow.com/questions/51084602/conversion-from-opencv-to-pil-is-not-accurate

```python
img_cv2 = cv2.imread(image_path)
img_cv2 = cv2.resize(
    img_cv2,
    (1024, 1024),
    interpolation=cv2.INTER_LINEAR,
)
img_cv2 = cv2.cvtColor(img_cv2, cv2.COLOR_BGR2RGB)/255
```

```python
img_pil = Image.open(image_path)
img_pil = img_pil.resize((1024, 1024), resample = Image.BILINEAR)
img_pil = np.asarray(img_pil).astype("float32") / 255
```

```python
def split_by_value(img: npt.NDArray, num_classes:int)-> npt.NDArray:
    height, width = img.shape[0], img.shape[1]
    split_img = np.zeros((height, width))
    split_img = np.expand_dims(split_img, axis=-1)
    for idx in range(1, num_classes):
        mask = img[:,:]==idx
        mask = np.expand_dims(mask, axis=-1)
        split_img = np.concatenate((split_img, mask), axis=-1)
        
    return split_img
```
